```python
#importação da biblioteca
import pandas as pd
```

```python
#leitura do arquivo json
dados = pd.read_json('/content/dados_hospedagem.json')
```

```python
#visualizar as primeiras 5 linhas
dados.head()
```

```python
#normalizar o json - tranformar em formato tabular
dados = pd.json_normalize(dados['info_moveis'])
dados
#observamos que temos dados numéricos, de texto e em lista
#observamos tb que a partir de descrição total tudo está em lista
```

```python
#verificar os nomes das colunas
colunas = list(dados.columns)
colunas
```

```python
#utilizando metodo explode, para retirar os dados que estão dentro de listas
#é utilizando um slice para fazer a transformação a partir do indice 3 até o final
dados = dados.explode(colunas[3:])
dados
```

```python
#resentando os indices para reordenar os dados
#o parametro inplace = True serve para substituir os dados e o drop = excluir os indices antigos
#observamos que temos numeros floats, inteiros e até como moeda
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
#verificando os tipos de dados
#object -> varios tipos de dados diferentes ou tipo especifico não especificado ou todos são strings
dados.info()
```

```python
import numpy as np
```

```python
#transformar a coluna max hospede para numero inteiro
#medoto astype permite que convertemos para um tipo especifico de dados

dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
#Verificando novamente os tipo
dados.info()
```

```python
#Fazendo novamente para outras colunas
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']

dados[col_numericas] = dados[col_numericas].astype(np.int64)

```

```python
#Verificando novamente os tipo
dados.info()
```

```python
#fazendo a modificação para numeros flutuantes
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
#Verificando novamente os tipo
dados.info()
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
#Agora sim podemos modificar para float, pois foram retirados virgulas e $
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
#Verificando novamente os tipo
dados.info()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
#Verificando novamente os tipo
dados.info()
```

```python
#Passando todo o texto para minusculo
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
dados['descricao_local'][100]
```

```python
#substitua através do regex que diz selecione tudo com exceção de texto de az(minusculo e maiusulo) e numeros de 0 a 9 e tb traço e aspas simples
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
#\veirifica pelo regex se tem caracteres antes e depois do traço, tipo se for um hifen solto ele pode ser excluido
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?<!\w)',' ', regex=True)
```

```python
#colocar o texto em lista
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
```
